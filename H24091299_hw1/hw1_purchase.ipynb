```python
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "C:/Users/User/Downloads/purchase_data.csv"
# Read purchasing file and store into pandas data frame
df= pd.read_csv(file_to_load)

a=df.groupby("SN")
d1f=pd.DataFrame({"Tatal Players":[len(a)]})
print(d1f)
```

```python
a=df["Price"].sum()
b=df.groupby("Item ID")
b=len(b)

c=df.shape[0]
d2f=pd.DataFrame({"Number of Unique Items":[b],"Average Price":["$"+str(round(a/c, 2))],"Number of Purchases":[c],"Total Revenue":["$"+str(a)]})
print(d2f)
```

```python
m=len(df.groupby("SN"))
a=df.groupby("Gender")
male=a.get_group("Male")
female=a.get_group("Female")
other=a.get_group("Other / Non-Disclosed")
x=[male,female,other]
p=[]
t=[]
for i in x:
    n=len(i.groupby("SN"))
    p.append(round(n/m*100,2))
    t.append(n)
d3f=pd.DataFrame({"Percentage of Players":p,"Total Count":t})
d3f.index=["Male","Female","Other / Non-Disclosed"]
print(d3f)
```

```python
a=df.groupby("Gender")
male=a.get_group("Male")
female=a.get_group("Female")
other=a.get_group("Other / Non-Disclosed")
x=[female,male,other]
pc=[]
app=[]
tpv=[]
aptpp=[]
for i in x:
    a=len(i)
    pc.append(a)
    b=i["Price"].sum()
    tpv.append(b)
    app.append(str(round(b/a,5)).rstrip("0"))
    n=len(i.groupby("SN"))
    aptpp.append("$"+str(round(b/n,2)))
d4f=pd.DataFrame({"Purchase Count":pc,"Average Purchase Price":app,"Total Purchase Value":tpv,"Avg Purchase Total per Person":aptpp})
d4f.index=["Feale","Male","Other / Non-Disclosed"]
d4f.index.name='Gender'
print(d4f)
```

```python
x10=df["Age"]<10
x15=df["Age"]<15
x20=df["Age"]<20
x25=df["Age"]<25
x30=df["Age"]<30
x35=df["Age"]<35
x40=df["Age"]<40
u10=df[df["Age"]<10]
u14=df[(x15&(x10==False))]
u19=df[(x20&(x15==False))]
u24=df[(x25&(x20==False))]
u29=df[(x30&(x25==False))]
u34=df[(x35&(x30==False))]
u39=df[(x40&(x35==False))]
u40=df[df["Age"]>=40]
x=[u10,u14,u19,u24,u29,u34,u39,u40]
pop=[]
tc=[]
t=len(df.groupby("SN"))
for i in x:
    a=len(i.groupby("SN"))
    tc.append(a)
    pop.append(round(a/t*100,2))
d5f=pd.DataFrame({"Percentage of Players":pop,"Total Count":tc})
d5f.index=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
print(d5f)
```

```python
a=df.groupby('SN')
b=[]
pc=[]
app=[]
tpv=[]
for i,j in a:
    b.append(i)
    aa=a.get_group(i)
    pc.append(len(aa))
    tpv.append(round(aa["Price"].sum(),2))
    app.append(round(aa["Price"].sum()/len(aa),2))
aaa=pd.DataFrame({'Purchase Count':pc,'Average Purchase Price':app,"Total Purchase Value":tpv})
aaa.index=b
d7f=aaa.sort_values(["Total Purchase Value"],ascending=False)
d7f=d7f.head()
for i in range(len(d7f)):
    d7f.iloc[i,2]="$"+str(d7f.iloc[i,2])
    d7f.iloc[i,1]="$"+str(d7f.iloc[i,1])
d7f.index.name='SN'
print(d7f)
```

```python
a=df.groupby(["Item ID","Item Name"])
b=[]
bb=[]
pc=[]
app=[]
tpv=[]
for i,j in a:
    b.append(i[0])
    bb.append(i[1])
    aa=a.get_group(i)
    pc.append(len(aa))
    tpv.append(round(aa["Price"].sum(),2))
    app.append(round(aa["Price"].sum()/len(aa),2))
aaa=pd.DataFrame({'Purchase Count':pc,'Average Purchase Price':app,"Total Purchase Value":tpv})
aaa.index=[b,bb]
aaa=aaa.rename_axis(("Item ID","Item Name"))
d8f=aaa.sort_values(['Purchase Count'],ascending=False)
d8f=d8f.head()
for i in range(len(d8f)):
    d8f.iloc[i,2]="$"+str(d8f.iloc[i,2])
    d8f.iloc[i,1]="$"+str(d8f.iloc[i,1])
print(d8f)
```
